- 문제1
데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후,
'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../Datasat/bigdata_certification_kr/basic1.csv")
df.head()
```

```python
df = df.sort_values('f5', ascending=False)
df.head(10)
```

```python
#'f5'컬럼 10개중 최소값
min = df['f5'][:10].min()
min
```

```python
df.iloc[:10,-1] = min
df.head(10)
```

```python
print(df[df['age']>=80]['f5'].mean() )
```

- 문제2
데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용해서,
'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고
두 표준편차 차이 계산하기

```python
df = pd.read_csv("../../Datasat/bigdata_certification_kr/basic1.csv")
df.head()
```

```python
data70, data30 = np.split(df, [int(.7*len(df))])
data70.tail()
```

```python
before_std = data70['f1'].std()
data70['f1'] = data70['f1'].fillna(data70['f1'].median())
after_std = data70['f1'].std()
print(abs(after_std - before_std))
```

- 문제3
데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오!
단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
df = pd.read_csv("../../Datasat/bigdata_certification_kr/basic1.csv")
df.head()
```

```python
std = df['age'].std() * 1.5
mean = df['age'].mean()
min_val = mean-std
max_val = mean+std
print(df[ (df['age']<min_val) | (df['age']>max_val)]['age'].sum())
```
