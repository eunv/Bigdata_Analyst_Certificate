basic1 데이터와 basic3 데이터를 'f4'값을 기준으로 병합하고,
병합한 데이터에서 r2결측치를 제거한다음, 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합을 구하시오

```python
import numpy as np
import pandas as pd
```

```python
b1 = pd.read_csv("../../Datasat/bigdata_certification_kr/basic1.csv")
b3 = pd.read_csv("../../Datasat/bigdata_certification_kr/basic3.csv")
b1.head()
```

```python
b3
```

```python
df = pd.merge(left=b1, right=b3, how="left", on="f4")
df.head()
```

```python
df.isnull().sum()
```

```python
print(df.shape)
df = df.dropna(subset=['r2'])
print(df.shape)
```

```python
df = df.reset_index()
```
