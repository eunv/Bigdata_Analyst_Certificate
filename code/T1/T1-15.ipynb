# f1의 결측치를 중앙값으로 채운다.
# 그리고 f4가 ISFJ와 f5가 20 이상인 
# f1의 평균값을 출력하시오

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../Datasat/bigdata_certification_kr/basic1.csv")
df.head()
```

```python
df = df.sort_values('age',ascending=False).head(20)
df
```

```python
df['f1'] = df['f1'].fillna(df['f1'].median())
```

```python
cond = (df['f4']=='ISFJ') & (df['f5'] >= 20)
cond
```

```python
print(df[cond]['f1'].mean())
```
